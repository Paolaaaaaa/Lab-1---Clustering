# 4. Implementación Clustering spectral

## 4.1 Preparación de los datos.

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.cluster import SpectralClustering
```

```python
name_cols_number = ['Time','Accident_Severity']
```

```python
label_encoder = LabelEncoder()#con label encoder
transformed_dt=dt_biciAlpes [name_cols_number].copy()
transformed_dt["Time"]= label_encoder.fit_transform(transformed_dt["Time"])
transformed_dt.describe()
```

```python
spClustering = clustering = SpectralClustering(n_clusters=3,
        assign_labels='discretize',
        random_state=0).fit(transformed_dt)
cols_number = transformed_dt.to_numpy()
spClustering.labels_
```

```python
plt.scatter(cols_number[spClustering.labels_ == 0, 0], cols_number[spClustering.labels_ == 0, 1], s = 100, marker='v', c = 'red', label = 'Cluster 1')
plt.scatter(cols_number[spClustering.labels_ == 1, 0], cols_number[spClustering.labels_ == 1, 1], s = 100, marker='o', c = 'blue', label = 'Cluster 2')

plt.title('Cluster of severity of accident')
plt.xlabel(name_cols_number[0])
plt.ylabel(name_cols_number[1])
plt.legend()
plt.show()
```
